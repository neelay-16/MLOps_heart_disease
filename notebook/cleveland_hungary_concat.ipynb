```python
import pandas as pd
import numpy as np

print("="*60)
print("DATASET EXPLORATION - CLEVELAND vs HUNGARIAN")
print("="*60)

# ====================== 1. CLEVELAND DATASET ======================
print("\n1. CLEVELAND DATASET")
print("-" * 40)

cleveland_path = r'C:\WILP 3rd sem\MLOps\Assignment_01\heart+disease\processed.cleveland.data'

col_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
             'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']

df_cleve = pd.read_csv(cleveland_path, header=None, names=col_names)           #header=None: Tells pandas that the first row is NOT column names, The file contains only raw data (numbers), no header row. names=col_names: Manually assigns the column names you provide in the list col_names, Without this, columns would be named 0, 1, 2, ..., 13, which is hard to work with.
df_cleve = df_cleve.replace('?', np.nan)                      #Replaces every '?' with NaN (Not a Number), which is pandas' standard way of representing missing data. In the UCI Heart Disease dataset, missing values are represented by the character '?' (question mark). Pandas treats '?' as a string, which creates problems later
df_cleve = df_cleve.apply(pd.to_numeric, errors='coerce')     #pd.to_numeric: Converts string/object columns into numeric type. errors='coerce': If any value cannot be converted to a number (e.g., some unexpected text), it will be converted to NaN instead of throwing an error. 'coerce' = "force conversion and turn bad values into NaN". Even after replacing '?', some columns might still be of object (string) type.

print(f"Shape: {df_cleve.shape}")
print("\nMissing Values:")
print(df_cleve.isnull().sum())
print("\nTarget Distribution:")
print(df_cleve['target'].value_counts().sort_index())
print("\nBasic Statistics:")
print(df_cleve.describe().round(2))
```

```python
# ====================== 2. HUNGARIAN DATASET ======================
print("\n\n2. HUNGARIAN DATASET")
print("-" * 40)

def load_hungarian(file_path):
    data = []
    current_patient = []
    
    with open(file_path, 'r') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            tokens = line.split()
            for token in tokens:
                token = token.strip()
                if token in ['-9', '-9.', 'name']:
                    if token == 'name':
                        continue
                    current_patient.append(np.nan)
                else:
                    try:
                        current_patient.append(float(token))
                    except ValueError:
                        continue
            if len(current_patient) >= 76:
                data.append(current_patient[:76])
                current_patient = []
    
    base_names = col_names[:]  
    extra_names = [f'attr_{i+1}' for i in range(14, 76)]
    df = pd.DataFrame(data, columns=base_names + extra_names)
    return df


hungarian_path = r'C:\WILP 3rd sem\MLOps\Assignment_01\heart+disease\hungarian.data'
df_hung = load_hungarian(hungarian_path)

print(f"Shape: {df_hung.shape}")
print("\nMissing Values (First 14 columns):")
print(df_hung.iloc[:, :14].isnull().sum())
print("\nTarget Distribution:")
print(df_hung['target'].value_counts().sort_index())
print("\nBasic Statistics (First 14 columns):")
print(df_hung.iloc[:, :14].describe().round(2))
```
